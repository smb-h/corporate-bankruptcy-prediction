```python
for fl_ratio in os.listdir(fin_ratio_dir):
    if fl_ratio.endswith('.xlsx'):
        this_df_ratio = pd.read_excel(os.path.join(fin_ratio_dir, fl_ratio))
        # drop cols with all NaNs
        this_df_ratio = this_df_ratio.dropna(axis=1, how='all')

        # this_df_ratio_copy = this_df_ratio.copy()
        # this_df_ratio_copy.columns = [col + "_" + fl_ratio.split(" ")[-1].split(".")[0] + '_ratio' if col != 'Company' else col for col in this_df_ratio.columns]
        
        this_df_ratio["year"] = fl_ratio.split(" ")[-1].split(".")[0]

        if df_ratio.empty:
            df_ratio = this_df_ratio
        else:
            # df_ratio = df_ratio.merge(this_df_ratio_copy, on='Company', how='outer')
            df_ratio = pd.concat([df_ratio, this_df_ratio], axis=0)
print(df_ratio.shape)
df_ratio.head()
```

```python
for fl_factor in os.listdir(fin_factors_dir):
    if fl_factor.endswith('.xlsx'):
        this_df_factors = pd.read_excel(os.path.join(fin_factors_dir, fl_factor))
        # drop last 2 cols
        this_df_ratio = this_df_ratio.iloc[:, :-2]
        # drop cols with all NaNs
        this_df_factors = this_df_factors.dropna(axis=1, how='all')
        this_df_factors.drop('Unnamed: 32', axis=1, inplace=True)

        # this_df_factors_copy = this_df_factors.copy()
        # this_df_factors_copy.columns = [col + "_" + fl_factor.split(" ")[-1].split(".")[0] + '_factor' if col != 'Company' else col for col in this_df_factors.columns]

        this_df_factors["year"] = fl_factor.split(" ")[-1].split(".")[0]

        if df_factors.empty:
            df_factors = this_df_factors
        else:
            # df_factors = pd.merge(df_factors, this_df_factors_copy, on='Company')
            df_factors = pd.concat([df_factors, this_df_factors], axis=0)
print(df_factors.shape)
df_factors.head()
```
